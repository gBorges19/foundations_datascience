```python
import pandas as pd

data = pd.read_csv('Happy.csv')
```

```python
happy = data['happyScore']
income = data['avg_income']
ineq = data['income_inequality'] #desigualdade
income.max()
```

```python
import matplotlib.pyplot as plt

plt.xlabel('income')
plt.ylabel('happy Score')
plt.scatter(income,happy, s=ineq*10, alpha=0.25) #tamanho = desigualdade
```
