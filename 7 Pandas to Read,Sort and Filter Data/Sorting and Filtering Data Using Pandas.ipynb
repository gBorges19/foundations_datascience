```python
import pandas as pd
import numpy as np

data = pd.read_csv('Happy.csv')
```

```python
data.sort_values('avg_income', inplace=True)

richest = data [ data ['avg_income'] > 15000 ]

#richest.iloc[0] #menor renda média dos paises ricos > 15
#richest.iloc[-1] #maior renda média dos paises ricos > 15
#richest.iloc[0:5] #os 5 primeiros de menor renda média > 15

all_mean = np.mean(data['avg_income']) # média total
rich_mean = np.mean(richest['avg_income']) # média entre os maiores que 15 mil

print(all_mean,rich_mean)
```
