```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches #para criar a elipse.

xs = [10, 100, 25, 67, 74]
ys = [125, 26, 66, 1, 10]

xys = [[10,125], [100,26], [25,66], [67,1], [74,10]]
xys = np.array(xys)

mean = np.mean(xys,0)

# calculando o desvio padrão.
std_dev = np.std(xys,0)

ellipse = patches.Ellipse([mean[0],mean[1]], std_dev[0]*2,std_dev[1]*2, alpha = 0.25)   # (largura,altura) (média,desvio padrão)

# alpha = value para determinar transparencia da elipse

# multiplos gráficos em um único gráfico, utilizando subparcelas
fig,graph = plt.subplots()

#plt.scatter(xys[:,0], xys[:,1])
#plt.scatter(mean[0], mean[1])

#to

graph.scatter(xys[:,0], xys[:,1])
graph.scatter(mean[0], mean[1])
graph.add_patch(ellipse)
```

```python
#Calcular a distancia entre um ponto

#xys[0] - mean --> é necessário calcular a distância euclidiana.  Soma, Quadrado, Raiz

dist1 = np.linalg.norm(xys[0] - mean)
dist2 = np.linalg.norm(xys[1] - mean)
dist3 = np.linalg.norm(xys[2] - mean)
dist4 = np.linalg.norm(xys[3] - mean)
dist5 = np.linalg.norm(xys[4] - mean)

print(dist1)
print(dist2)
print(dist3)
print(dist4)
print(dist5)
```
