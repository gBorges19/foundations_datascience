```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans

data = pd.read_csv('vgsales.csv')

best =  data[data['Rank'] < 20]

genre = best['Genre']
sales = best['Global_Sales']
rank = best['Rank']
name = best['Name']

plt.text(best.iloc[0]['Global_Sales'],
         best.iloc[0]['Genre'],
         best.iloc[0]['Name'])

plt.text(best.iloc[9]['Global_Sales'],
         best.iloc[9]['Genre'],
         best.iloc[9]['Name'])

plt.text(best.iloc[-1]['Global_Sales'],
         best.iloc[-1]['Genre'],
         best.iloc[-1]['Name'])

sales_rank = np.column_stack((sales,rank))
                             
km_res = KMeans(n_clusters=2).fit(sales_rank)

clusters = km_res.cluster_centers_

plt.scatter(sales,genre,s=100,alpha=0.25)
plt.scatter(clusters[:,0],clusters[:,1], s=150)
```
